```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_sas("/kaggle/input/bankloan/bankloan .sas7bdat")
```

```python
data
```

```python
data.info()
```

```python
df = data[~data["default"].isnull()]
```

```python
df
```

```python
def IQR(x):
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    iqr = q3 - q1
    lf = q1 - (1.5*iqr)
    uf = q3+(1.5*iqr)
    print("LF",lf)
    print("UF",uf)
```

```python
df["employ"] = np.where(df["employ"]>25.5,25.5,df["employ"])

df["address"] = np.where(df["address"]>25.5,25.5,df["address"])

df["income"] = np.where(df["income"]>101.5,101.5,df["income"])

df["debtinc"] = np.where(df["debtinc"]>27.812,27.812,df["debtinc"])

df["creddebt"] = np.where(df["creddebt"]>4.20,4.20,df["creddebt"])

df["othdebt"] = np.where(df["othdebt"]>8.24,8.24,df["othdebt"])

df["ed"] = np.where(df["ed"]>3.5,3.5,df["ed"])
```

```python
df.columns
```

```python
X = df[['age', 'ed', 'employ', 'address', 'income', 'debtinc', 'creddebt',
       'othdebt']]
y = df["default"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=55)
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
print("Train accuracy",log.score(X_train,y_train))
print("Test accuracy",log.score(X_test,y_test))
```

***Decision Tree***

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(random_state = 55)
```

```python
dt.fit(X_train,y_train)
```

```python
print("Train :",dt.score(X_train,y_train))
print("Test :",dt.score(X_test,y_test))
```

***Grid Search for Decision Tree***

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': [None, 3, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt']
}
```

```python
grid_search = GridSearchCV(dt, param_grid, cv=5, n_jobs=2, scoring='accuracy')
```

```python
grid_search.fit(X_train,y_train)
```

```python
print("Best parameters found:", grid_search.best_params_)
print("Best cross-validation accuracy:", grid_search.best_score_)
```

```python
print("Train :", grid_search.score(X_train,y_train))
print("Test :",grid_search.score(X_test,y_test))
```

```python
# classification report

y_pred = grid_search.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

***Random Forest***

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestClassifier(random_state=42)
```

***Grid Search for Random Forest***

```python
param_grid = {
    'n_estimators': [50, 100],           # Number of trees
    'max_depth': [None, 5, 10],          # Depth of each tree
    'min_samples_split': [2, 5],         # When to split a node
    'min_samples_leaf': [1, 2],          # Minimum samples per leaf
    'max_features': ['sqrt', 'log2'],    # Feature selection
}
```

```python
grid_rf = GridSearchCV(rf, param_grid, cv=5, n_jobs=-1, scoring='f1')  

grid_rf.fit(X_train, y_train)

```

```python
y_pred_rf = grid_rf.predict(X_test)
print("Best Parameters:", grid_rf.best_params_)
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
```
